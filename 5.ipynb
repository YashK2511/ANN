```python
import numpy as np
```

```python
x = np.array([
    [1, 1, 1],
    [-1, -1, 1]
])

y = np.array([
    [1, 1, -1, -1],
    [-1, -1, 1, 1]
])
```

```python
W = np.dot(x.T, y)
print("W:\n", W)
```

```python
def sign(x):
    return np.where(x >= 0, 1, -1)
```

```python
y1_out = sign(np.dot(x[0], W))
y2_out = sign(np.dot(x[1], W))

print("\nForward Output:")
print("y1:", y1_out)
print("y2:", y2_out)
```

```python
x1_out = sign(np.dot(y1_out, W.T))
x2_out = sign(np.dot(y2_out, W.T))

print("\nBackward Output:")
print("x1:", x1_out)
print("x2:", x2_out)
```
