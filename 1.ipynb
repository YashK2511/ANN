```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.linspace(-10,10,20)
y = sigmoid(x)
y1 = tanh(x)
y2 = relu(x)
```

```python
plt.figure(figsize=(15,10))
plt.title("Activation functions",fontsize = 20)

plt.subplot(2,2,1)
plt.plot(x,y,color = "blue",marker = "o",label = "sigmoid")
plt.title("Sigmoid")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.grid()

plt.subplot(2,2,2)
plt.plot(x,y1,color = "red",marker = "o",label = "tanh")
plt.title("tanh")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.grid()

plt.subplot(2,2,3)
plt.plot(x,y2,color = "green",marker = "o",label = "relu")
plt.title("relu")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.grid()
```
