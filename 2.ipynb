```python
import numpy as np
```

```python
x1 = [0,0,1,1]
x2 = [0,1,0,1]
```

```python
def mcp(x1,x2):
    w1 = 1
    w2 = -1
    theta = 1
    output = []
    for i in range(len(x1)):
        y = np.dot([w1,w2],[x1[i],x2[i]])
        output.append(1) if y >= theta else output.append(0)
    return output
```

```python
output = mcp(x1,x2)
```

```python
for i in range(len(x1)):
    print("x1 = ",x1[i]," x2 = ",x2[i]," output = ",output[i])
```
