```python
import numpy as np
import numpy.random as ar
import matplotlib.pyplot as plt
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
t = np.array([0,0,0,1])
```

```python
w = ar.rand(2)
b = ar.rand(1)
lr = 0.1

for epoch in range(50):
    e = 0
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        yin = 1 if z >= 0 else 0
        error = t[i] - yin
        
        if error != 0:
            e += 1
            w += lr * error * X[i]
            b += lr * error
    
    if e == 0:
        break

print("Final weights:", w)
print("Final bias:", b)
```

```python
yin = []
for i in range(len(X)):
    z = np.dot(X[i], w) + b
    yin.append(1 if z >= 0 else 0)

plt.figure(figsize=(6,6))
for i in range(len(X)):
    X1 = [0,0,1,1]
    X2 = [0,1,0,1]
    print(X1[i],X2[i],yin[i])
plt.scatter(X1, X2, c=t, cmap="coolwarm", edgecolors="k", s=100)

x1 = np.linspace(-1, 2, 100)
x2 = -(x1 * w[0] + b) / w[1]

plt.plot(x1, x2, c="black", label="Decision boundary")

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron Decision Boundary for AND Gate")
plt.legend()
plt.grid()
plt.show()
```
