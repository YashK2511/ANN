```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
x = np.array([[1],[2],[3],[4]])
y = np.array([[0],[0],[1],[1]])
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1,activation = 'sigmoid')
])
```

```python
model.compile(
    optimizer = "sgd",
    loss = "binary_crossentropy",
    metrics = ["accuracy"]
)
```

```python
model.fit(x,y,epochs=10)
```

```python
n=2
predictions = model.predict(np.array([[n]]))
print(predictions)
if predictions > 0.5:
    print("pass")
else:
    print("fail")
```
